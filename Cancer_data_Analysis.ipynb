```python
#rendon forest for trget severty score

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score

#converting categorical colom to numerical colom
categrical_colom = ["Gender" , "Country_Region" ,"Cancer_Type" , "Cancer_Stage"]
for col in categrical_colom:
  le=LabelEncoder()
  data[col]=le.fit_transform(data[col])



# preapering feature and input
X= data.drop(columns =["Patient_ID" , "Survival_Years" ,"Target_Severity_Score" ,"Treatment_Cost_USD" ])
y_severity= data["Survival_Years"]


#train test split
X_train_s , X_test_s , Y_train_s , Y_test_s = train_test_split(X ,y_severity, test_size =0.2, random_state=40 )

param_grid = {
    'n_estimators' : [100,200],
    'max_depth' :[5,10,None],
    'min_samples_split' : [2,5],
    'min_samples_leaf' : [1,2]
}

#train the model
model = RandomForestRegressor( random_state=40)
GSC = GridSearchCV(model ,param_grid , cv = 3 , scoring='r2' , n_jobs = -1)
GSC.fit(X_train_s , Y_train_s)

best_rt_severity = GSC.best_estimator_


# evaluate the model
train_r2_severity= r2_score(Y_train_s , best_rt_severity.predict(X_train_s))
test_r2_severity= r2_score(Y_test_s , best_rt_severity.predict(X_test_s))



```

#asses treatment higher treatment cost is assosiated with longer survival

null hypo (H.) = thre is no coretion bt treatment cost and sirvival year

alternate hypo (h1) = thre is  coretion(+ or -) bt treatment cost and sirvival year

```python
from  scipy.stats import pearsonr, spearmanr
x = data["Treatment_Cost_USD"]
y = data["Survival_Years"]

#performing pearson corretion test
pearson_corr , pearson_p = pearsonr(x ,y)
print(f"person corretion coffi : {pearson_corr}")
print(f"person p_ value : {pearson_p}")

#spermen corretrion test
spearman_corr , spearman_p = spearmanr(x ,y)
print(f"spearman  corretion coffi : {spearman_corr}")
print(f"spearman p_ value : {spearman_p}")


alpha = 0.5

def interpret_corr(corr , p ,method):
  if p < alpha:
    print(f"{method} , we reject null hypo")
  else:
      print(f"{method} , we fail to reject null hypo")

interpret_corr(pearson_corr,pearson_p , "Pearson")
interpret_corr(spearman_corr,spearman_p , "Spearman")
```

```python
stage_order = ["Stage 0" , "Stage I" ,"Stage II" , "Stage III" , "Stage IV"]
group_state =  df.groupby("Cancer_Stage")[["Treatment_Cost_USD","Survival_Years"]].mean().reset_index()
```

treatment cost vs cencer stage

null hypo (H.) = avg treatment treatment cost is the same all cencer stages

alternate hypo (h1) = at lest one stages has a different avg cost

survival yeras vs cencer steges
null hypo (H.) = avg treatment servival year is the same all cencer stages

alternate hypo (h1) = at lest one stages has a different servival duretion

```python
grouped_cost=[]
grouped_servival=[]

for stage in stage_order:
  stage_data = df[df["Cancer_Stage"] == stage]
  cost=stage_data["Treatment_Cost_USD"]
  survival = stage_data["Survival_Years"]
  grouped_cost.append(cost)
  grouped_servival.append(survival)
```

```python
len(grouped_cost)
```

```python
# chek for normality
from scipy.stats import shapiro , f_oneway
normal_cost = 0;
normal_servival = 0;


for i  in range (len(stage_order)):
  cost_p =  shapiro(grouped_cost[i]).pvalue
  surv_p = shapiro(grouped_servival[i]).pvalue
  print(f"cost {cost_p} for group  {i}")
  print(f"servival {surv_p} for group  {i}")
  if cost_p<0.05:
    normal_cost+=1
  if surv_p<0.05:
    normal_servival+=1
print(cost_p)
print(surv_p)
```
